### Requirement FR1.1 - Find the arithmetic mean



```python
# add code here
# Read the Glocuse data from task1.dat
task1 = [i.strip() for i in open("task1.dat").readlines()]

# coverting the string into integers
data_task1 = (list(map(int, task1)))

# finding the mean
mean = sum(data_task1) / len(data_task1)

print("There are a total of", len(data_task1), "observations in the Glocuse data.")
print("The arithmetic mean of Glocuse is", mean,".")
```

### Requirement FR1.2 - Find the standard deviation

```python
# add code here
# computing for the variance and standard deviation
variance = sum([((x - mean) ** 2) for x in data_task1]) / len(data_task1)
std = variance ** 0.5
print("The standard deviation of Glocuse is", std)
```

### Requirement FR1.3 - Find the min/max values 

```python
# add code here
# Maximum and Minimum values of data
max(data_task1)
min(data_task1)

Min = min(data_task1)
Max = max(data_task1)
print("The Minimum value of Glocuse is", Min,"and the Maximum value is",Max)
```
